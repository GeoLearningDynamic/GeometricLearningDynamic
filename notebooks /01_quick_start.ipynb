# Geometric Learning Dynamics - Quick Start

This notebook provides a quick introduction to the geometric learning dynamics framework.

```python
import sys
sys.path.append('..')

from geometric_learning.data import SphereDataset
from geometric_learning.models import GatedNetwork
from geometric_learning.simulator import GradientFlowSimulator
import matplotlib.pyplot as plt

# Create dataset
dataset = SphereDataset(n_samples=500, dim=10, seed=42)

# Create network
net = GatedNetwork(M=50, dim=10, seed=42)

# Run simulation
sim = GradientFlowSimulator(dataset, net, lr=0.1)
sim.run(steps=500, record_every=10)

# Plot results
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.semilogy(sim.history['step'], sim.history['norm_perp'].mean(axis=1))
plt.xlabel('Step')
plt.ylabel('||w⊥||')
plt.title('Manifold Alignment')

plt.subplot(1,2,2)
plt.plot(sim.history['step'], sim.history['L12'])
plt.xlabel('Step')
plt.ylabel('L₁₂')
plt.title('Angular Momentum')
plt.tight_layout()
plt.show()
```
